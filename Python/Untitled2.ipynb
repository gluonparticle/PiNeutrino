```python
def is_palindrome(number):
    number_str = str(number)
    return number_str == number_str[::-1]

def count_digits(number):
    digit_count = [0] * 10
    while number > 0:
        digit = number % 10
        digit_count[digit] += 1
        number //= 10
    return digit_count

try:
    num = int(input("Enter a number: "))


    if num < 0:
        print("Please enter a non-negative number.")
    else:
        if is_palindrome(num):
            print(f"{num} is a palindrome.")
        else:
            print(f"{num} is not a palindrome.")

        digit_count = count_digits(num)


        for i in range(10):
            if digit_count[i] > 0:
                print(f"Digit {i} appears {digit_count[i]} times in the number.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def fn(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fn(n-2) + fn(n-1)

try:
    num = int(input("Enter a number: "))
    if num > 0:
        print(f"fn({num}) = {fn(num)}")
    else:
        print("Error in input")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def BinToDec(x):
    dec = 0
    i = 0
    while x > 0:
        r = x % 10
        if r != 0 and r != 1:
            print("Enter a valid Binary number")
            return 0
        else:
            dec = dec + r * 2**i
            x = x // 10
            i += 1
    return dec

def OctaToHexa(n):
    num = n
    dec = 0
    base = 1
    temp = num
    while temp:
        r = temp % 10
        temp = temp // 10
        dec += r * base
        base = base * 8

    # Convert decimal to hexadecimal
    result = ''
    while dec != 0:
        temp = dec % 16
        if temp < 10:
            result = str(temp) + result
        else:
            result = chr(temp + 55) + result
        dec = dec // 16
    return result

try:
    # Binary to Decimal Conversion
    x = int(input("Enter a Binary number: "))
    result = BinToDec(x)
    if result:
        print(f"The Decimal equivalent of {x} is {result}")

    # Octal to Hexadecimal Conversion
    y = int(input("Enter an Octal number: "))
    result = OctaToHexa(y)
    if result:
        print(f"The Hexadecimal equivalent of {y} is {result}")

except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
def string_similarity(x, y):
    # Strip spaces from both strings
    x = x.strip()
    y = y.strip()

    # Find the minimum length of both strings to compare up to that length
    min_length = min(len(x), len(y))

    # Initialize similarity counter
    sim = 0

    # Compare characters from both strings
    for i in range(min_length):
        if x[i] == y[i]:
            sim += 1

    # Calculate the similarity ratio based on the longer string length
    similarity = sim / max(len(x), len(y))

    return similarity

# Input from user
x = input("Enter the first string: ")
y = input("Enter the second string: ")

# Calculate and display the similarity
similarity = string_similarity(x, y)
print(f"The similarity between the two given strings is {similarity:.2f}")
```
